```python
def bin_to_dec(n):
  if float(n).is_integer():
    return int(str(n), 2)

  integer, decimal = str(n).split('.')
  res = int(str(integer), 2)

  expoent = -1
  dec_value = 0
  for d in decimal:
    dec_value += int(d)*2**expoent
    expoent -=1
  res += dec_value
  return res
```

```python
def dec_to_bin(n):
  if float(n).is_integer():
    return bin(n).replace("0b","")

  integer, decimal = str(n).split('.')
  res = bin(int(integer)).replace("0b","") + '.'

  x = float('0.' + decimal)
  for i in range(10):
    x *= 2
    res += str(int(x))
    x %= 1

  return res
```

## 4) Seja um sistema de aritmética de ponto flutuante de quatro dígitos e base decimal.
```python
x = 0.7237e4
y = 0.2145e-3
z = 0.2585e-1
```

### a) x + y + z

```python
xy = round(x + y)
xyz = round(xy + z)
er_xy = .5 * 10**(-3)
er_xyz = er_xy + .5 * 10**(-3)

print(f'x + y = {xy}')
print(f'x + y + z = {xyz}')
print(f'Erro relativo de x + y = {er_xy}')
print(f'Erro relativo de x + y + z = {er_xyz}')
```

### b) x - y - z

```python
xy = round(x - y)
xyz = round(xy - z)
er_xy = .5 * 10**(-3)
er_xyz = er_xy + .5 * 10**(-3)

print(f'x - y = {xy}')
print(f'x - y - z = {xyz}')
print(f'Erro relativo de x - y = {er_xy}')
print(f'Erro relativo de x - y - z = {er_xyz}')
```

```python
xy = round(x / y)
er_xy = .5 * 10**(-3)

print(f'x * y = {xy}')
print(f'Erro relativo de x 8 y = {er_xy}')
```

### d) (xy)/z

```python
xy = round(x*y, 3)
xyz = round(xy / z, 2)
er_xy = .5 * 10**(-3)
er_xyz = er_xy + .5 * 10**(-3)

print(f'x * y = {xy}')
print(f'(x * y) / z = {xyz}')
print(f'Erro relativo de x * y = {er_xy}')
print(f'Erro relativo de (x * y) / z = {er_xyz}')
```

## 5) Calcule o polinômio $y = x^{3} - 7x^{2} + 8x - 0,35$ em $x = 1,37$. Use aritmética com 3 algarismos significativos e truncamento. Calcule o erro relativo percentual.

```python
x = 1.37
y = round(x**3 - 7*x**2 + 8*x - 0.35, 4)
x2 = float(str(x*x)[:4])
x3 = round(x2 * x, 2)
x27 = round(7 * x2)
x8 = float(str(8*x)[:4])
y_calc = float(str(x3 - x27 + x8 - 0.35)[:4])
erro_relativo = round(abs((y - y_calc) / y_calc), 3)

print(f'y nominal = {y}')
print(f'y calculado = {y_calc:.3f}')
print(f'Erro relativo = {erro_relativo}')
```

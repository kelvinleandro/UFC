## **2 ‚Äì A corrente I em amp√®res e a tens√£o V em volts em uma l√¢mpada se relacionam conforme a seguinte express√£o:**

**I = (V/V<sub>0</sub>)<sup>ùú∂</sup>, em que Œ± √© um n√∫mero entre 0 e 1, e V<sub>0</sub> √© tens√£o nominal em volts. Sabendo que V<sub>0</sub> = 220 ¬± 3% e Œ± = -0,8 ¬± 4%, calcule a corrente e o erro relativo associado quando a tens√£o vale 220 ¬± 1%.**

OBS: Este problema pode ser resolvido de duas formas distintas: usando a express√£o aproximada para a propaga√ß√£o do erro e inspecionando os valores m√°ximos e m√≠nimos que a express√£o pode assumir. Pratique os dois m√©todos. Dica: lembre que x<sup>Œ±</sup> = e<sup>Œ±.ln(x)</sup>

```python
v, vo = 220, 220
a = -0.8
i = (v/vo)**a
print(i)
```

```python
vo_max = vo + vo*0.03
vo_min = vo - vo*0.03
v_max = v + v*0.01
v_min = v - v*0.01
a_max = a + a*0.04
a_min = a - a*0.04

i_max = (v_max/vo_max)**a_max
i_min = (v_min/vo_min)**a_min

print(i_max, i_min)
```

```python
print(abs(i - i_max))
print(abs(i - i_min))
```

## **3 - Considere as express√µes:**

$$\frac{\exp(\frac{1}{\mu})}{1 + \exp(\frac{1}{\mu})}$$ e

$$\frac{1}{\exp(\frac{-1}{\mu}) + 1}$$

Com ùúá > 0. Verifique que elas s√£o id√™nticas como fun√ß√µes reais. Teste no computador cada uma delas para ùúá = 0,1, ùúá = 0,01, ùúá = 0,001. Qual dessas express√µes √© mais adequada quando ùúá √© um n√∫mero pequeno? Por qu√™?

**Resposta:** Verificando as duas express√µes abaixo, percebe-se que a segunda express√£o √© mais adequada, pois a primeira op√ß√£o ao colocar um n√∫mero muito pequeno como 0.001 ocorreu erro de *range error* durante a execu√ß√£o.

```python
from math import exp
```

```python
def exp_1(mu):
  return exp(1/mu) / (1 + exp(1/mu))

def exp_2(mu):
  return 1 / (exp(-1/mu) + 1)
```

```python
# mu = 0.1
print(exp_1(0.1), exp_2(0.1))
```

```python
# mu = 0.01
print(exp_1(0.01), exp_2(0.01))
```

```python
# mu = 0.001
print(exp_1(0.0001), exp_2(0.001))
```

## **2 – A corrente I em ampères e a tensão V em volts em uma lâmpada se relacionam conforme a seguinte expressão:**

**I = (V/V<sub>0</sub>)<sup>𝜶</sup>, em que α é um número entre 0 e 1, e V<sub>0</sub> é tensão nominal em volts. Sabendo que V<sub>0</sub> = 220 ± 3% e α = -0,8 ± 4%, calcule a corrente e o erro relativo associado quando a tensão vale 220 ± 1%.**

OBS: Este problema pode ser resolvido de duas formas distintas: usando a expressão aproximada para a propagação do erro e inspecionando os valores máximos e mínimos que a expressão pode assumir. Pratique os dois métodos. Dica: lembre que x<sup>α</sup> = e<sup>α.ln(x)</sup>

```python
v, vo = 220, 220
a = -0.8
i = (v/vo)**a
print(i)
```

```python
vo_max = vo + vo*0.03
vo_min = vo - vo*0.03
v_max = v + v*0.01
v_min = v - v*0.01
a_max = a + a*0.04
a_min = a - a*0.04

i_max = (v_max/vo_max)**a_max
i_min = (v_min/vo_min)**a_min

print(i_max, i_min)
```

```python
print(abs(i - i_max))
print(abs(i - i_min))
```

## **3 - Considere as expressões:**

$$\frac{\exp(\frac{1}{\mu})}{1 + \exp(\frac{1}{\mu})}$$ e

$$\frac{1}{\exp(\frac{-1}{\mu}) + 1}$$

Com 𝜇 > 0. Verifique que elas são idênticas como funções reais. Teste no computador cada uma delas para 𝜇 = 0,1, 𝜇 = 0,01, 𝜇 = 0,001. Qual dessas expressões é mais adequada quando 𝜇 é um número pequeno? Por quê?

**Resposta:** Verificando as duas expressões abaixo, percebe-se que a segunda expressão é mais adequada, pois a primeira opção ao colocar um número muito pequeno como 0.001 ocorreu erro de *range error* durante a execução.

```python
from math import exp
```

```python
def exp_1(mu):
  return exp(1/mu) / (1 + exp(1/mu))

def exp_2(mu):
  return 1 / (exp(-1/mu) + 1)
```

```python
# mu = 0.1
print(exp_1(0.1), exp_2(0.1))
```

```python
# mu = 0.01
print(exp_1(0.01), exp_2(0.01))
```

```python
# mu = 0.001
print(exp_1(0.0001), exp_2(0.001))
```

```python
import cv2
import cv2.typing as cvt
import numpy as np
from math import log
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
def reduzir_niveis_intensidade(imagem, niveis):
    # Verifica se o número de níveis é uma potência de 2
    if (niveis & (niveis - 1)) != 0:
        raise ValueError("O número de níveis deve ser uma potência de 2.")

    # Calcula o fator de redução de intensidade
    fator_reducao = 256 // niveis

    # Aplica a redução de intensidade
    imagem_reduzida = (imagem // fator_reducao) * fator_reducao
    return imagem_reduzida
```

```python
# Caminho para a imagem
caminho_imagem = 'drive/MyDrive/PDI/assets/Fig2.21(a).jpg'

# Carrega a imagem em escala de cinza
imagem = cv2.imread(caminho_imagem, cv2.IMREAD_GRAYSCALE)
if imagem is None:
    print("Erro ao carregar a imagem!")
    exit()

# Define o número de níveis de intensidade desejado
niveis_desejados = 16  # Você pode alterar para 2, 4, 8, 16, etc.

# Reduz a imagem para o número desejado de níveis de intensidade
imagem_reduzida = reduzir_niveis_intensidade(imagem, niveis_desejados)

# Exibe a imagem original e a imagem reduzida
cv2_imshow(imagem_reduzida)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(10, 5))
axes = axes.flatten()

for i, ax in enumerate(axes):
    level = 2 ** (i + 1)
    ax.imshow(reduzir_niveis_intensidade(imagem, level), cmap='gray')
    ax.set_title(f'Níveis: {level}')
    ax.axis('off')
```
